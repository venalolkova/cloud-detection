```python
class CloudData(torch_data.Dataset):
    def __init__(self, X, y):
        super(CloudData, self).__init__()
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X[idx],self.y[idx]
```

```python
class MyFirstAE(nn.Module):
    
    def __init__(self, in_channel, out_channel):
        super(MyFirstAE, self).__init__()
#         self.threshold = 0.9
        self.conv_encode1 = self.contracting_block(in_channels=in_channel,mid_channel=16, out_channels=16)
        self.conv_maxpool1 = torch.nn.MaxPool2d(kernel_size=2)
        self.conv_encode2 = self.contracting_block(in_channels=16, mid_channel=64, out_channels=64)
        self.conv_maxpool2 = torch.nn.MaxPool2d(kernel_size=2)
        self.conv_encode3 = self.contracting_block(in_channels=64, mid_channel=128, out_channels=128)
        self.conv_maxpool3 = torch.nn.MaxPool2d(kernel_size=2)
        self.conv_encode4 = self.contracting_block(in_channels=128, mid_channel=256, out_channels=256)
        self.conv_maxpool4 = torch.nn.MaxPool2d(kernel_size=2)
        self.conv_encode5 = self.contracting_block(in_channels=256, mid_channel=512, out_channels=512)
        self.conv_maxpool5 = torch.nn.MaxPool2d(kernel_size=2)

        # Bottleneck
        self.bottleneck = torch.nn.Sequential(
                            torch.nn.Conv2d(kernel_size=3, in_channels=512, out_channels=1024,padding=1),
                            torch.nn.BatchNorm2d(1024),
                            torch.nn.ReLU(),
#                             torch.nn.BatchNorm2d(1024),
                            torch.nn.Conv2d(kernel_size=3, in_channels=1024, out_channels=1024,padding=1),
                            torch.nn.BatchNorm2d(1024),
                            torch.nn.ReLU(),
#                             torch.nn.BatchNorm2d(1024),
                            torch.nn.ConvTranspose2d(in_channels=1024, out_channels=512, kernel_size=3, stride=2, padding=1, output_padding=1,dilation=1)
                            )
        # Decode
        self.conv_decode1 = self.expansive_block(1024, 512, 256) #first argument have size(decode_block[1]+copy_block[1])
        self.conv_decode2 = self.expansive_block(512, 256, 128)
        self.conv_decode3 = self.expansive_block(256, 128, 64)
        self.conv_decode4 = self.expansive_block(128, 32, 16)
        self.final_layer = self.final_block(32, 16, out_channel) 

    def contracting_block( self, in_channels, mid_channel,out_channels, kernel_size=3,padding=1):
        mid_channel = out_channels
        block = torch.nn.Sequential(
                    torch.nn.Conv2d(kernel_size=kernel_size, in_channels=in_channels, out_channels=mid_channel,padding=padding),
                    torch.nn.BatchNorm2d(mid_channel),
                    torch.nn.ReLU(),
#                     torch.nn.BatchNorm2d(mid_channel),
                    torch.nn.Conv2d(kernel_size=kernel_size, in_channels=mid_channel, out_channels=out_channels,padding=padding),
                    torch.nn.BatchNorm2d(out_channels),
                    torch.nn.ReLU(),
#                     torch.nn.BatchNorm2d(out_channels),
                )
        return block

    def expansive_block( self, in_channels, mid_channel,out_channels, kernel_size=3,padding=1):
        block = torch.nn.Sequential(
                torch.nn.Conv2d(kernel_size=kernel_size, in_channels=in_channels, out_channels=mid_channel,padding=padding),
                torch.nn.BatchNorm2d(mid_channel),
                torch.nn.ReLU(),
#                 torch.nn.BatchNorm2d(mid_channel),
                torch.nn.Conv2d(kernel_size=kernel_size, in_channels=mid_channel, out_channels=mid_channel,padding=padding),
                torch.nn.BatchNorm2d(mid_channel),
                torch.nn.ReLU(),
#                 torch.nn.BatchNorm2d(mid_channel),
                torch.nn.ConvTranspose2d(in_channels=mid_channel, out_channels=out_channels, kernel_size=3, stride=2, padding=1, output_padding=1,dilation=1)
                )
        return  block

    def final_block( self, in_channels, mid_channel, out_channels, kernel_size=3,padding=1):
        block = torch.nn.Sequential(
                torch.nn.Conv2d(kernel_size=kernel_size, in_channels=in_channels, out_channels=mid_channel,padding=padding),
                torch.nn.BatchNorm2d(mid_channel),
                torch.nn.ReLU(),
#                 torch.nn.BatchNorm2d(mid_channel),
                torch.nn.Conv2d(kernel_size=kernel_size, in_channels=mid_channel, out_channels=mid_channel,padding=padding),
                torch.nn.BatchNorm2d(mid_channel),
                torch.nn.ReLU(),
#                 torch.nn.BatchNorm2d(mid_channel),
                torch.nn.Conv2d(kernel_size=kernel_size, in_channels=mid_channel, out_channels=out_channels, padding=padding),
                torch.nn.BatchNorm2d(out_channels),
                torch.nn.Sigmoid(),
#                 torch.nn.BatchNorm2d(out_channels),
                )
        return  block
    
    def forward(self, x):
        # Encode
        encode_block1 = self.conv_encode1(x)
        encode_pool1 = self.conv_maxpool1(encode_block1)
        encode_block2 = self.conv_encode2(encode_pool1)
        encode_pool2 = self.conv_maxpool2(encode_block2)
        encode_block3 = self.conv_encode3(encode_pool2)
        encode_pool3 = self.conv_maxpool3(encode_block3)
        encode_block4 = self.conv_encode4(encode_pool3)
        encode_pool4 = self.conv_maxpool4(encode_block4)
        encode_block5 = self.conv_encode5(encode_pool4)
        encode_pool5 = self.conv_maxpool5(encode_block5)
        # Bottleneck
        bottleneck1 = self.bottleneck(encode_pool5)
        # Decode
        copy_block1 = torch.cat((bottleneck1, encode_block5), 1)
        decode_block1 = self.conv_decode1(copy_block1)

        copy_block2 = torch.cat((decode_block1, encode_block4), 1)
        decode_block2 = self.conv_decode2(copy_block2)

        copy_block3 = torch.cat((decode_block2, encode_block3), 1)
        decode_block3 = self.conv_decode3(copy_block3)
        
        copy_block4 = torch.cat((decode_block3, encode_block2), 1)
        decode_block4 = self.conv_decode4(copy_block4)
        
        copy_block5 = torch.cat((decode_block4, encode_block1), 1)

        probability_output = self.final_layer(copy_block5)

#         ones = torch.ones(probability_output.size())
#         zeroes = torch.zeros(probability_output.size())

#         output = torch.where(probability_output<self.threshold,zeroes,ones)
        
#         output.requires_grad_(True)
        
        return  probability_output
```

```python
def jaccard_loss(pred,true,epsilon=0.0001):
    pred = pred.flatten()
    true = true.flatten()
    assert true.size() == pred.size()
    a = torch.sum(true*pred)
    b = torch.sum(true) + torch.sum(pred) - a 
    return -(a+epsilon)/(b+epsilon)
```

```python
def train(epochs, net, criterion, optimizer, train_loader, val_loader,scheduler=None, verbose=True, save_dir=None):
    net.to(device)
    for epoch in range(1, epochs+1):
        net.train()
        for X, y in train_loader:
            # Perform one step of minibatch stochastic gradient descent
            
            nn_outputs = net.forward(X)
            loss = criterion(nn_outputs, y)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
        # define NN evaluation, i.e. turn off dropouts, batchnorms, etc.
        net.eval()
        for X, y in val_loader:
            # Compute the validation loss
            
            nn_outputs = net.forward(X)
            val_loss = criterion(nn_outputs, y)
         
        if scheduler is not None:
            scheduler.step()
        freq = max(epochs//20,1)
#         if verbose and epoch%freq==0:
        print('Epoch {}/{} || Loss:  Train {:.4f} | Validation {:.4f}'.format(epoch, epochs, loss.item(), val_loss.item()))
```

```python
torch.manual_seed(42)

device = 'cpu'

net = MyFirstAE(in_channel=4, out_channel=1)  
criterion = jaccard_loss
optimizer = torch.optim.Adam(net.parameters(),lr = 0.001)

Lambda = lambda epoch: 0.9 ** epoch
scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda=Lambda)


train_loader_raw = torch_data.DataLoader(train_dset_raw, batch_size=1, shuffle=False) 
val_loader_raw = torch_data.DataLoader(val_dset_raw, batch_size=1, shuffle=False) 

train_loader_cor = torch_data.DataLoader(train_dset_cor, batch_size=1, shuffle=False) 
val_loader_cor = torch_data.DataLoader(val_dset_cor, batch_size=1, shuffle=False) 
```

```python
%%time
torch.manual_seed(42)
train(20, net, criterion, optimizer, train_loader_raw, val_loader_raw, scheduler)
```
